```python
import random

import json
with open ("E:\DEPI\AI_Data_Science_R4\GitHub_Repo\DEPI-Rahman\src\Python\Session_1\Code\data.json", "r") as file:

    responses = json.load (file)

name = input ('Hi, please enter your name...')
while True:
    user_input = input (f"ChatBot: How can i assist you {name}...?")
    user_response = False
    for key in responses:
        if key in user_input:
            print (f"ChatBot: {random.choice (responses [key])}")
            user_response = True
            break
    if not user_response:
        print (f"ChatBot: {random.choice (responses ['default'])}")
    if user_input == "goodbye":
        print (f"ChatBot: {random.choice (responses ['goodbye'])}")
        break
```
