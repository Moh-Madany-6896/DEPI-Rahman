```python
def recurs (num : int):

    if num == 0:
        return 1
    if num == 1:
        return 1
    
    return num * recurs (num - 1)

recurs (5)
```

```python
def is_prime (num : int):
    for x in range (2, num):
        if num % x == 0:
            return False
        return True
    
is_prime (20)
```
