```python
def recurs (num : int):

    if num == 0:
        return 1
    if num == 1:
        return 1
    
    return num * recurs (num - 1)

recurs (5)
```

```python
def is_prime (num : int):
    for x in range (2, num):
        if num % x == 0:
            return False
        return True
    
is_prime (20)
```

```python
def dividors (num1, num2):
    mn = min (num1, num2)
    common = []
    for x in range (1, mn):
        if num1 % x == 0 and num2 % x == 0:
            common.append (x)
    return (common)

dividors (100, 50)
```

```python
def even (num):
    result = []

    for i in range (num + 1):
        if i % 2 == 0:
            result.append (i)
    return (result)

even (20)
```

```python
def encode (x : str):
    sp = x.split (" ")
    result = " ".join (sp [::-1])
    return (result)


encode ("My Name Is Mohamed")
```

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass
    
    @abstractmethod
    def stop(self):
        pass
    
    def get_info(self):
        return "This is a vehicle"

# Concrete class implementing the abstract class
class Car(Vehicle):
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
    
    # Must implement all abstract methods
    def start(self):
        return f"{self.brand} {self.model}: Engine started!"
    
    def stop(self):
        return f"{self.brand} {self.model}: Engine stopped!"
    
    # Can add own methods
    def drive(self):
        return f"Driving {self.brand} {self.model}"

# Another concrete class
class Bicycle(Vehicle):
    def start(self):
        return "Pedaling started!"
    
    def stop(self):
        return "Bicycle stopped!"

# Using the classes
car = Car("Toyota", "Camry")
print(car.start())  # Output: Toyota Camry: Engine started!
print(car.drive())  # Output: Driving Toyota Camry

bike = Bicycle()
print(bike.start())  # Output: Pedaling started!
print(bike.get_info())  # Output: This is a vehicle

# Can't instantiate abstract class
# vehicle = Vehicle()  # TypeError!
```
