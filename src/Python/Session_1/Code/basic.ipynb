```python
def recurs (num : int):

    if num == 0:
        return 1
    if num == 1:
        return 1
    
    return num * recurs (num - 1)

recurs (5)
```

```python
def is_prime (num : int):
    for x in range (2, num):
        if num % x == 0:
            return False
        return True
    
is_prime (20)
```

```python
def dividors (num1, num2):
    mn = min (num1, num2)
    common = []
    for x in range (1, mn):
        if num1 % x == 0 and num2 % x == 0:
            common.append (x)
    return (common)

dividors (100, 50)
```

```python
def even (num):
    result = []

    for i in range (num + 1):
        if i % 2 == 0:
            result.append (i)
    return (result)

even (20)
```

```python
def encode (x : str):
    sp = x.split (" ")
    result = " ".join (sp [::-1])
    return (result)


encode ("My Name Is Mohamed")
```

```python
def gen_pass (x):
    import random
    import string
    password = ""
    charts1 = string.ascii_letters
    charts2 = string.digits + string.hexdigits
    for _ in range(x + 1):
        password += random.choice (charts1)
    return password

gen_pass (10)
```

```python
def unique (x : list):
    unique_words = set (x)
    print (unique_words)
    for i in unique_words:
        print (f"{i}: {x.count (i)}")

unique (["mohamed", 29, 35, "depi", "Depi", "Mohamed", "mohamed"])
```
